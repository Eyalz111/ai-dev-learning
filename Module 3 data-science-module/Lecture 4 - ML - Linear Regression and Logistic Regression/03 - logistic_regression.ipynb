# Logistic Regression

## Part 1 - Data Preprocessing

### Importing the dataset

```python
import pandas as pd
dataset = pd.read_csv('log_data.csv')
```

```python
dataset.head(10)
```

### Getting the inputs and output

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

### Creating the Training Set and the Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

## Part 2 - Building and training the model

### Building the model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0)
```

### Training the model

```python
model.fit(X_train, y_train)
```

### Inference

Making the predictions of the data points in the test set

```python
y_pred = model.predict(sc.transform(X_test))
```

```python
y_pred
```

## Part 3: Evaluating the model

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

### Accuracy

```python
(84+47)/(84+47+3+3)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# Calculate metrics manually using confusion matrix values
# From confusion matrix [[84, 3], [3, 47]]
TP = 47  # True Positives (correctly predicted malignant)
FP = 3   # False Positives (incorrectly predicted malignant)
FN = 3   # False Negatives (incorrectly predicted benign)
TN = 84  # True Negatives (correctly predicted benign)

# Calculate Recall = TP / (TP + FN)
recall = TP / (TP + FN)
print("Recall:", recall)

# Calculate Precision = TP / (TP + FP) 
precision = TP / (TP + FP)
print("Precision:", precision)

# Calculate F1 Score = 2 * (Precision * Recall) / (Precision + Recall)
f1 = 2 * (precision * recall) / (precision + recall)
print("F1 Score:", f1)
```
