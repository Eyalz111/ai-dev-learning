```python
# 4. How can you calculate the correlation matrix for the DataFrame 
# to understand relationships between numerical features?
# use housing_df.select_dtypes(include=['number']) to select only numerical columns

import seaborn as sns
import matplotlib.pyplot as plt

# Create correlation matrix
corr_matrix = housing_df.select_dtypes(include=['number']).corr()

# Get top 20 features most correlated with SalePrice
top_20_features = corr_matrix['SalePrice'].sort_values(ascending=False).head(20).index
corr_matrix_top20 = housing_df[top_20_features].corr()

# Create heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix_top20, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap (Top 20 Features)')
plt.tight_layout()
plt.show()
```

## Column Creation


```python
# Create a Boolean column 'is_expensive' where True if 'SalePrice' is greater than 200000


housing_df['is_expensive'] = np.where(housing_df['SalePrice'] > 200000, "is expensive", "not that expensive")

housing_df[['SalePrice', 'is_expensive']]
```

```python
# Create a new column 'tax_amount' where if 'is_expensive' is True, calculate tax as 0.1 * 'SalePrice', otherwise 0

housing_df['tax_amount'] = np.where(housing_df['is_expensive'] == "is expensive", 
                                  housing_df['SalePrice'] * 0.1,
                                  0) # Added missing second argument

housing_df[['SalePrice', 'is_expensive','tax_amount']]
                                  
```
